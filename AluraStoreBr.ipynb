### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja_1 = pd.read_csv(url)
loja_2 = pd.read_csv(url2)
loja_3 = pd.read_csv(url3)
loja_4 = pd.read_csv(url4)
```

```python
# Adicionar a coluna de 'Lucro' ao dataset
# Como não foi especificado se o frete é pago pelo comprador ou pela loja, para o Lucro foi considerado que o frete é arcado pela loja.
loja_1['Lucro'] = loja_1['Preço'] - loja_1['Frete']
loja_2['Lucro'] = loja_2['Preço'] - loja_2['Frete']
loja_3['Lucro'] = loja_3['Preço'] - loja_3['Frete']
loja_4['Lucro'] = loja_4['Preço'] - loja_4['Frete']
```

```python
# Calcular o lucro médio de cada uma das 4 lojas
lucro_loja_1 = loja_1['Lucro'].mean()
lucro_loja_2 = loja_2['Lucro'].mean()
lucro_loja_3 = loja_3["Lucro"].mean()
lucro_loja_4 = loja_4['Lucro'].mean()

print(f'O lucro médio da Loja 1 foi de: {lucro_loja_1}')
print(f'O lucro médio da Loja 2 foi de: {lucro_loja_2}')
print(f'O lucro médio da Loja 3 foi de: {lucro_loja_3}')
print(f'O lucro médio da Loja 4 foi de: {lucro_loja_4}')
```

```python
# Podemos criar um gráfico paa visualizar a média anual de vendas de cada loja
loja_1['Ano'] = pd.to_datetime(loja_1['Data da Compra'], format='%d/%m/%Y').dt.year
loja_2['Ano'] = pd.to_datetime(loja_2['Data da Compra'], format='%d/%m/%Y').dt.year
loja_3['Ano'] = pd.to_datetime(loja_3['Data da Compra'], format='%d/%m/%Y').dt.year
loja_4['Ano'] = pd.to_datetime(loja_4['Data da Compra'], format='%d/%m/%Y').dt.year
```

```python
loja_1
```

```python
lucro_anual_loja_1 = loja_1.groupby('Ano')['Lucro'].mean()
lucro_anual_loja_2 = loja_2.groupby('Ano')['Lucro'].mean()
lucro_anual_loja_3 = loja_3.groupby('Ano')['Lucro'].mean()
lucro_anual_loja_4 = loja_4.groupby('Ano')['Lucro'].mean()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Transformar as Series em DataFrames e adicionar uma coluna 'Loja'
df_lucro_anual_loja_1 = lucro_anual_loja_1.reset_index().assign(Loja='Loja 1')
df_lucro_anual_loja_2 = lucro_anual_loja_2.reset_index().assign(Loja='Loja 2')
df_lucro_anual_loja_3 = lucro_anual_loja_3.reset_index().assign(Loja='Loja 3')
df_lucro_anual_loja_4 = lucro_anual_loja_4.reset_index().assign(Loja='Loja 4')

# Concatenar os DataFrames
df_lucro_anual = pd.concat([df_lucro_anual_loja_1, df_lucro_anual_loja_2,
                            df_lucro_anual_loja_3, df_lucro_anual_loja_4])

# Criar o gráfico
plt.figure(figsize=(10, 6))
sns.barplot(x='Ano', y='Lucro', hue='Loja', data=df_lucro_anual)
plt.title('Média Anual de Lucro por Loja')
plt.xlabel('Ano')
plt.ylabel('Lucro Médio')
plt.legend(title='Loja')
plt.show()
```

```python
# Quais foram os vendedores com maior número de vendas em cada loja?

# Agrupar por vendedor e contar o número de vendas
vendas_por_vendedor_loja_1 = loja_1.groupby('Vendedor').size()
vendas_por_vendedor_loja_2 = loja_2.groupby('Vendedor').size()
vendas_por_vendedor_loja_3 = loja_3.groupby('Vendedor').size()
vendas_por_vendedor_loja_4 = loja_4.groupby('Vendedor').size()

# Ordenar os vendedores pelo número de vendas em ordem decrescente
vendas_por_vendedor_loja_1 = vendas_por_vendedor_loja_1.sort_values(ascending=False)
vendas_por_vendedor_loja_2 = vendas_por_vendedor_loja_2.sort_values(ascending=False)
vendas_por_vendedor_loja_3 = vendas_por_vendedor_loja_1.sort_values(ascending=False)
vendas_por_vendedor_loja_4 = vendas_por_vendedor_loja_2.sort_values(ascending=False)

# Imprimir os vendedores com maior número de vendas
print('Os vendedores da Loja 1 com maior quantidade de vendas são: ')
print(vendas_por_vendedor_loja_1)

print('Os vendedores da Loja 2 com maior quantiade de vendas são: ')
print(vendas_por_vendedor_loja_2)

print('Os vendedores da Loja 3 com maior quantiade de vendas são: ')
print(vendas_por_vendedor_loja_3)

print('Os vendedores da Loja 4 com maior quantiade de vendas são: ')
print(vendas_por_vendedor_loja_4)
```

```python
# Transformar as Series em DataFrames e adicionar uma coluna 'Loja'
df_vendas_loja_1 = vendas_por_vendedor_loja_1.reset_index().assign(Loja='Loja 1')
df_vendas_loja_2 = vendas_por_vendedor_loja_2.reset_index().assign(Loja='Loja 2')
df_vendas_loja_3 = vendas_por_vendedor_loja_3.reset_index().assign(Loja='Loja 3')
df_vendas_loja_4 = vendas_por_vendedor_loja_4.reset_index().assign(Loja='Loja 4')

# Concatenar os DataFrames
df_vendas = pd.concat([df_vendas_loja_1, df_vendas_loja_2, df_vendas_loja_3, df_vendas_loja_4])

# Renomear a coluna 0 para Vendas
df_vendas = df_vendas.rename(columns={0: 'Vendas'})
```

```python
# Criar o gráfico
plt.figure(figsize=(12, 8))  # Ajuste o tamanho da figura conforme necessário
sns.barplot(y='Vendedor', x='Vendas', hue='Loja', data=df_vendas, orient='h')
plt.title('Vendas por Vendedor e Loja')
plt.ylabel('Vendedor')
plt.xlabel('Número de Vendas')
plt.legend(title='Loja')
plt.show()
```

#1. Análise do faturamento


```python
# 1 - calcular o faturamento das 4 lojas somando a coluna 'Preço' de cada uma das lojas
faturamento_loja_1 = loja_1['Preço'].sum()
faturamento_loja_2 = loja_2['Preço'].sum()
faturamento_loja_3 = loja_3['Preço'].sum()
faturamento_loja_4 = loja_4['Preço'].sum()
```

```python
print(f'O faturamento total da Loja 1 foi de: {faturamento_loja_1}')
print(f'O faturamento total da Loja 2 foi de: {faturamento_loja_2}')
print(f'O faturamento total da Loja 3 foi de: {faturamento_loja_3}')
print(f'O faturamento total da Loja 4 foi de: {faturamento_loja_4}')
```

```python
#Exibir gráfico com faturamento total de cada loja
import matplotlib.pyplot
import seaborn as sns

#dicionário com os dados de faturamento
faturamento_lojas = {
    'Loja 1': faturamento_loja_1,
    'Loja 2': faturamento_loja_2,
    'Loja 3': faturamento_loja_3,
    'Loja 4': faturamento_loja_4
}

#criar o gráfico de barras
plt.figure(figsize=(8, 6))
sns.barplot(x=list(faturamento_lojas.keys()), y=list(faturamento_lojas.values()))
plt.title('Faturamento Total por Loja')
plt.xlabel('Loja')
plt.ylabel('Faturamento')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Criar um DataFrame com os dados de faturamento
faturamento_df = pd.DataFrame.from_dict(faturamento_lojas, orient='index', columns=['Faturamento'])
faturamento_df['Loja'] = faturamento_df.index

# Plotar o gráfico de linhas
sns.lineplot(x='Loja', y='Faturamento', data=faturamento_df, marker='o')
plt.title('Faturamento por Loja')
plt.xlabel('Loja')
plt.ylabel('Faturamento (R$)')
plt.xticks(rotation=45, ha='right')  # Rotacionar os rótulos do eixo x para melhor visualização
plt.tight_layout()  # Ajustar o layout para evitar sobreposição de elementos
plt.show()
```
